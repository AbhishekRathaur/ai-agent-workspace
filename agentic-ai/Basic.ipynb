```python
from typing import TypedDict, Annotated
import operator

# 1. Define the Shared Memory Schema
class MemoryState(TypedDict):
    input_text: str
    processed_steps: Annotated[list, operator.add]

# 2. Define the Workers
def uppercase_worker(state: MemoryState):
    new_text = state["input_text"].upper()
    return {"input_text": new_text, "processed_steps": ["Converted to uppercase"]}

def dramatic_worker(state: MemoryState):
    final_text = f"💥 {state['input_text']} 💥"
    return {"input_text": final_text, "processed_steps": ["Added drama"]}

print("✅ Definitions registered in Python memory!")
```

```python
from langgraph.graph import StateGraph, START, END

# 1. Initialize the map with our memory schema
builder = StateGraph(MemoryState)

# 2. Register our workers on the map
builder.add_node("upper_node", uppercase_worker)
builder.add_node("drama_node", dramatic_worker)

# 3. Draw the layout lines (Edges)
builder.add_edge(START, "upper_node")       # Start point -> Worker 1
builder.add_edge("upper_node", "drama_node") # Worker 1 -> Worker 2
builder.add_edge("drama_node", END)          # Worker 2 -> End point

# 4. Compile the map into an executable application
app = builder.compile()
print("🎉 Graph compiled successfully!")
```

```python
# 1. Create the initial data packet (must match the MemoryState keys)
raw_materials = {"input_text": "hello cursor mastery"}

print("🚀 Pushing data into the graph...")

# 2. Fire the data into your compiled app
final_state = app.invoke(raw_materials)

print("\n--- EXECUTION COMPLETE ---")
print(f"Final Text: {final_state['input_text']}")
print(f"History Log: {final_state['processed_steps']}")
```

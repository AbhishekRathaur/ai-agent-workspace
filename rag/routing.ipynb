```python
import litellm
import json

# Define our production execution paths
def handle_chitchat(query: str):
    print(f"-> [ROUTED TO CHITCHAT HANDLER]")
    # Pass directly to LLM without hitting any databases
    messages = [{"role": "user", "content": query}]
    response = litellm.completion(model="ollama/llama3.2", messages=messages, api_base="http://localhost:11434")
    return response.choices[0].message.content

def handle_rag_search(query: str):
    print(f"-> [ROUTED TO KNOWLEDGE BASE RAG ENGINE]")
    return f"Mock RAG Output: Found specific architecture documents matching '{query}'."

def handle_system_command(query: str):
    print(f"-> [ROUTED TO HARDWARE SYSTEM COMMAND EXECUTION]")
    return "Mock System Output: Securely purged session logs and reset connection state."

print("✓ Cell 1: Routing handlers and destination functions fully declared.")
```

```python
def semantic_router(user_query: str) -> str:
    # Build a strict classification prompt map
    prompt = [
        {
            "role": "system",
            "content": (
                "You are an isolated, high-speed network routing switchboard. Your single job is to classify "
                "the intent of the user's input string into exactly one of three routing keys.\n\n"
                "Your routing keys are:\n"
                "1. 'CHITCHAT' -> General greetings, casual talk, jokes, or non-technical prompts.\n"
                "2. 'RAG_SEARCH' -> Highly specific technical questions about software architecture, ports, configurations, or data frameworks.\n"
                "3. 'SYSTEM_COMMAND' -> Requests to clear history, wipe state, restart systems, or purge logs.\n\n"
                "Output your response ONLY as a single raw string containing the route key name. Do not include spaces, markdown, punctuation, or explanations."
            )
        },
        {
            "role": "user",
            "content": f"Input to route: {user_query}"
        }
    ]
    
    response = litellm.completion(
        model="ollama/llama3.2",
        messages=prompt,
        api_base="http://localhost:11434",
        temperature=0.0  # Force maximal determinism
    )
    
    # Strip away any accidental whitespace or markdown wrapping from the model
    route_key = response.choices[0].message.content.strip().upper()
    if "CHITCHAT" in route_key: return "CHITCHAT"
    if "RAG_SEARCH" in route_key: return "RAG_SEARCH"
    if "SYSTEM_COMMAND" in route_key: return "SYSTEM_COMMAND"
    
    return "CHITCHAT"  # Default fallback safeguard

print("✓ Cell 2: Semantic routing controller successfully compiled.")
```

```python
def process_incoming_request(user_input: str):
    print(f"\nIncoming Query: '{user_input}'")
    
    # Step A: Run classification pass
    determined_route = semantic_router(user_input)
    print(f"Identified Route Class: {determined_route}")
    
    # Step B: Execute the structural conditional switch block
    if determined_route == "CHITCHAT":
        output = handle_chitchat(user_input)
    elif determined_route == "RAG_SEARCH":
        output = handle_rag_search(user_input)
    elif determined_route == "SYSTEM_COMMAND":
        output = handle_system_command(user_input)
    else:
        output = "Error: Routing failure."
        
    print(f"Response Matrix Result: {output}\n" + "-"*50)

print("✓ Cell 3: Orchestration wrapper bound.")
```

```python
# Test 1: Casual ChitChat
process_incoming_request("Good morning assistant! How are you doing today?")

```
